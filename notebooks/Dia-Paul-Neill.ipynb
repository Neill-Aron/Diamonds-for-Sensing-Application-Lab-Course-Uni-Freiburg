```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter

# === Lorentz-Dip-Form (negativ) ===
def lorentz_dip(f, A, f0, gamma):
    return -A * (gamma/2)**2 / ((f - f0)**2 + (gamma/2)**2)

# === Modell: Summe von n Lorentz-Dips + Offset ===
def multi_lorentz_dip(f, *params):
    offset = params[-1]
    n = (len(params) - 1) // 3
    result = np.full_like(f, offset)
    for i in range(n):
        A = params[3*i]
        f0 = params[3*i+1]
        gamma = params[3*i+2]
        result += lorentz_dip(f, A, f0, gamma)
    return result

# === Daten vorbereiten ===
f = df["Frequenz_Hz"].values
y = df["Signal"].values
y_smooth = savgol_filter(y, window_length=11, polyorder=3)

# === Anzahl der Dips ===
n_dips = 10

# === Startwerte schätzen ===
f_min, f_max = f.min(), f.max()
f0_guesses = np.linspace(f_min + 0.05, f_max - 0.05, n_dips)
initial_guess = []

for f0 in f0_guesses:
    initial_guess += [0.0002, f0, 0.05]  # A, f0, gamma

initial_guess += [0.0005]  # offset

# === Fit durchführen ===
popt, pcov = curve_fit(multi_lorentz_dip, f, y_smooth, p0=initial_guess, maxfev=10000)
y_fit = multi_lorentz_dip(f, *popt)

# === Plot ===
plt.figure(figsize=(10,6))
plt.plot(f, y, label="Rohdaten", color="lightgray")
plt.plot(f, y_smooth, label="Savitzky-Golay-glättung", color="navy")
plt.plot(f, y_fit, label=f"{n_dips}-facher Lorentz-Fit", color="crimson", lw=2)

plt.title(f"ODMR-Spektrum mit {n_dips} Lorentz-Dips (nach unten)", fontsize=14)
plt.xlabel("Frequenz [Hz]", fontsize=12)
plt.ylabel("Photolumineszenz [a.u.]", fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# === Fit-Ergebnisse ausgeben ===
for i in range(n_dips):
    A, f0, gamma = popt[3*i], popt[3*i+1], popt[3*i+2]
    print(f"Dip {i+1}: f0 = {f0:.6f} Hz, Γ = {gamma:.6f} Hz, Tiefe = {A:.6f}")
```
